# Principal Component Analysis (PCA)

The goal to dimensionality reduction is to replace a larger set of correlated variables with a smaller set of derived variables and loose the minimum amount of information. The best way of minimizing the loss of information is preservation of variance.

PCA is a data reduction technique that transforms a larger number of correlated variables into a much smaller set of correlated variables called as *Principal Components*. In other words PCA is a method of extracting important variables (in form of components) from a large set of variables available in a dataset.

The main idea of PCA is to reduce the dimensionality of dataset consisting of many variables correlated with each other, while retaining maximum information (variance) in the dataset. 

The features derived are low dimensional in nature. These components aim to capture as much information as possible with high explained variance.

The first component has the highest variance followed by second, third and so on. *The components must be uncorrelated*

PCA is applied on a dataset with numeric variables.

PCA is a statistical procedure that transforms and converts a dataset into a new dataset containing linearly uncorrelated variables, known as principal components.

```python
# Principal Component analysis
# Importing necessary packages and loading the dataset
from sklearn.decomposition import PCA
from sklearn.datasets import make_friedman1
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

x, y = make_friedman1()
```

```python
x
```

```python
y
```

```python
# Performing PCA
pca = PCA(n_components = 4)
pca.fit(x)

x_transformed = pca.transform(x)

print("Original shape of Friedman dataset: ", x.shape)
print("Reduced shape of Friedman dataset: ", x_transformed.shape)
print("Shape of Principal Components: ", pca.components_.shape)
print("Details of Principal Components: \n", pca.components_.round(3))
print("Component-wise explained variance: \n", pca.explained_variance_ratio_.round(3))
```

```python
# Creating the chart for depicting components
plt.matshow(pca.components_)
plt.yticks(range(4), ["First component", "Second component", "Third component", "Fourth comonent"])
plt.colorbar()
plt.xticks(range(10))
plt.tight_layout()
plt.xlabel("Features of Friedman dataset")
plt.ylabel("Principal Components")
plt.show()
```
